## Inline Question 2:

AdaGrad, like Adam, is a per-parameter optimization method that uses the following update rule:

```
cache += dw**2
w += - learning_rate * dw / (np.sqrt(cache) + eps)
```

John notices that when he was training a network with AdaGrad that the updates became very small, and that his network was learning slowly. Using your knowledge of the AdaGrad update rule, why do you think the updates would become very small? Would Adam have the same issue?


## Answer:
[FILL THIS IN]

**Answer：**

在 AdaGrad 中，每个参数都会维护一个累积的梯度平方项 `cache`，并且这个 `cache` 会随着训练过程**单调递增**，因为每一步都会把 `dw**2` 加进去。随着训练迭代次数增加，`cache` 的数值会越来越大，从而使得更新公式中的分母
$\sqrt{\text{cache}} + \epsilon$
不断增大。结果就是**有效学习率**
$\frac{learning_rate}{\sqrt{cache}+\epsilon}$
会越来越小，导致参数更新幅度逐渐减小，最终更新几乎停止，网络学习速度显著变慢。这正是 John 观察到“更新变得很小、学习变慢”的原因。

Adam 通常不会有同样严重的问题。虽然 Adam 也使用了梯度平方的二阶矩信息，但它并不是简单地累加所有历史梯度平方，而是通过指数滑动平均（带衰减因子 $\beta_2$）来估计梯度的二阶矩，这使得较早的梯度影响会逐渐减弱。此外，Adam 还对一阶矩和二阶矩都进行了偏置修正，使得学习率在训练早期和后期都能保持在一个相对合理的尺度。因此，与 AdaGrad 相比，Adam 不会因为训练时间变长而导致学习率持续衰减到几乎为零，通常在深度网络训练中表现得更加稳定和高效。
